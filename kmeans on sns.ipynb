```python
sn = range(1,11);print(sn)
kmeans_sns = [KMeans(n_clusters = i) for i in sn]
kmeans_sns
score = [kmeans_sns[i].fit(sns).score(sns) * -1 for i in range(len(kmeans_sns))]
print(score)
```

```python
import matplotlib.pyplot as plt
```
